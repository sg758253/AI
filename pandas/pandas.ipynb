# FillNa

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df1.head(10)
```

```python
#fill by scaler value (single value)
df1.fillna(0)
```

```python
#fill by scaler value on each specific column
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df1.fillna({'parental level of education':'some college', 'lunch':'standard', 'writing score':2},inplace=True)
df1.head(10)
```

```python
#fill next value
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df1.fillna(inplace=True,method='ffill')
df1.head(10)
```

```python
#fill next value
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df1.fillna(inplace=True,method='pad')
df1.head(10)
```

```python
#fill back value
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df1.fillna(inplace=True,method='bfill')
df1.head(10)
```

```python
#error: Must specify a fill 'value' or 'method'.
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df1.fillna(axis=0)
```

```python
# 
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df2=df1.fillna(axis=1, method='ffill')
df2.head(10)
```

```python
# on same dataframe, due to datatype values can't be change 
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df1.fillna(inplace=True,axis=1, method='ffill')
df1.head(10)
```

```python
# limit should greate than zero, value or method must be specify
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df2=df1.fillna(0,limit=1)
df2.head(10)
```

```python
# limit should greate than zero. value or method must be specify
df1 = pd.read_csv('https://raw.githubusercontent.com/sg758253/AI/main/pandas/StudentsPerformance.csv')
df2=df1.fillna(method='ffill',limit=2)
df2.head(10)
```
